## Look into NLTK

### Get data

```python
from bs4 import BeautifulSoup
import requests

url = "https://en.wikipedia.org/wiki/Wiki"

html = requests.get(url).content

soup = BeautifulSoup(html,'lxml')
data = soup.find_all('p')
paragraphs = ''.join(data[i].get_text(strip=True) for i in range(0, len(data)))
```

### Split words

```python
tokens = [tok for tok in paragraphs.split()]
print("Total number of tokens: " + str(len(tokens)))
```

#### First 100 tokens

```python
print(tokens[:100])
```

### Frequency Detection

```python
import nltk

Freq_dist = nltk.FreqDist(tokens)
print(Freq_dist)
```

#### Frequencies

```python
for k, v in Freq_dist.items():
    print(str(k) + ':' + str(v))
```

#### Frequency Distributions Plot

```python
Freq_dist.plot(50, cumulative=False)
```

### Delete stop words

```python
stopwords = ['the', 'a', 'an', 'is', 'are', 'to', 'at', 'and', 'or', 'that', 'can', 'where', 'what', 'how', 'of']
clean_tokens = [tok for tok in tokens if len(tok.lower()) > 1 and (tok.lower() not in stopwords)]
```

### Plot the distributions

```python
Freq_dist = nltk.FreqDist(clean_tokens)
Freq_dist.plot(50, cumulative=False)
```
